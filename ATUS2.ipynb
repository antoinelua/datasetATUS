```python
import pandas as pd
```

```python
url2019 = 'https://raw.githubusercontent.com/antoinelua/datasetATUS/main/atus_anual_2019.csv'
url2020 = 'https://raw.githubusercontent.com/antoinelua/datasetATUS/main/atus_anual_2020.csv'
url2021 = 'https://raw.githubusercontent.com/antoinelua/datasetATUS/main/atus_anual_2021.csv'
url2022 = 'https://raw.githubusercontent.com/antoinelua/datasetATUS/main/atus_anual_2022.csv'
```

```python
atus2019 = pd.read_csv(url2019)
atus2020 = pd.read_csv(url2020)
atus2021 = pd.read_csv(url2021)
atus2022 = pd.read_csv(url2022)
```

```python
atus2022['FECHA'] = pd.to_datetime(atus2022[['ANIO', 'MES', 'ID_DIA', 'ID_HORA', 'ID_MINUTO']].astype(str).agg('-'.join, axis=1), format='%Y-%m-%d-%H-%M')
```

```python
#  Agregue una columna para Área cuyos valores serán Urbana o Suburbana dependiendo del área donde ocurrió el accidente.

atus2022['ÁREA'] = 'Suburbana'

atus2022.loc[atus2022['URBANA'] == 'Accidente en intersección', 'ÁREA'] = 'Urbana'

atus2022
```

```python
atus2022 = atus2022.drop(columns=['Área'])
```

```python
atus2022
```

```python
atus2022 = atus2022.drop(['URBANA', 'SUBURBANA'], axis=1)
```

```python
atus2022
```

```python
atus2022['FECHA_MES'] = pd.to_datetime(atus2022[['ANIO', 'MES']].astype(str).agg('-'.join, axis=1))

meses_sin_accidentes = atus2022.groupby(['ID_ENTIDAD', 'ID_MUNICIPIO'])['FECHA_MES'].nunique().reset_index()

municipio_mas_seguro_por_estado = meses_sin_accidentes.loc[meses_sin_accidentes.groupby('ID_ENTIDAD')['FECHA_MES'].idxmin()]

print("Municipio más seguro por estado:")
municipio_mas_seguro_por_estado

```

```python
ultimo_anio = atus2022['ANIO'].max()

accidentes_ultimo_anio = atus2022[atus2022['ANIO'] == ultimo_anio]

# Agrupar los datos por tipo de accidente y contar cuántos hay de cada tipo
resumen_accidentes = accidentes_ultimo_anio.groupby('TIPACCID').size().reset_index(name='COUNT')

# Calcular el porcentaje que cada tipo de accidente representa del total de accidentes del último año
total_accidentes = resumen_accidentes['COUNT'].sum()
resumen_accidentes['PORCENTAJE'] = (resumen_accidentes['COUNT'] / total_accidentes) * 100

print("Resumen de accidentes del último año por tipo:")
resumen_accidentes
```
